## **Feature Encoding**

Coverting Categorical Column into Numerical Column

#### gender

```python
# gender

data['Gender'].value_counts()
```

```python
# Create a new numerical column
data['Gender_Num'] = data['Gender'].map({'Male': 1, 'Female': 0})
```

```python
data.head()
```

```python
data = data.drop(columns=['Gender'])
```

```python
data.head()
```

#### EducationBackground

```python
data['EducationBackground'].value_counts()
```

```python
data['Education'] = data['EducationBackground'].map({'Life Sciences': 1,'Medical': 2, 'Marketing': 3,'Technical Degree': 4,'Other': 5,'Human Resources': 6})
```

```python
data.head()
```

```python
data = data.drop(columns=['EducationBackground'])
```

```python
data.head()
```

#### MaritalStatus

```python
data['MaritalStatus'].value_counts()
```

```python
data['MaritalStatus_num'] = data['MaritalStatus'].map({'Married': 2, 'Single': 1, 'Divorced': 0})
```

```python
data = data.drop(columns=['MaritalStatus'])
```

```python
data.head()
```

```python
data = data.drop(columns=['EmpNumber'])
```

#### EmpDepartment	

```python
data['EmpDepartment'].value_counts()
```

```python
data['Department'] = data['EmpDepartment'].map({'Sales': 0,'Development': 1,'Research & Development': 2,'Human Resources': 3,'Finance': 4,'Data Science': 5})
```

```python
data = data.drop(columns=['EmpDepartment'])
```
