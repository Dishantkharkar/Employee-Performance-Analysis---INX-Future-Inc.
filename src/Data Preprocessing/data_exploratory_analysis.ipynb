## **Univariate Analysis**

```python
plt.figure(figsize=(25,20),facecolor="white")
plotnumber=1

for column in data:
  if plotnumber<27:
    ax=plt.subplot(7,4,plotnumber)
    sns.histplot(x=data[column])
    plt.xlabel(column,fontsize=10)

  plotnumber+=1
plt.tight_layout()
```

#### **Insights**
* **Age**: 
   - from the above graph we see that there are so many employee whose age  range between 25 to 40
   - Maximum age is 60

* **Gender**:
   - From the graph we see that male employees are more than female employee

* **EducationBackground**:
   - In the campany, there are high number of employee which have Life sciene background and after that Medical background

* **MaritalStatus**:
  - In the campany, higher number of employee have a Marital status, after that single and lower number is Divorced
